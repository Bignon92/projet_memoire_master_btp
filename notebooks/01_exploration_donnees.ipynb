## 4. Détection des problèmes

- valeurs manquantes ;
- doublons ;
- incohérences.

```python
#Valeurs manquantes
print("===== VALEURS MANQUANTES PARC AUTO =====")
display(parc_auto.isnull().sum())

print("===== VALEURS MANQUANTES SINISTRE AUTO =====")
display(sinistre_auto.isnull().sum())

print("===== VALEURS MANQUANTES SINISTRE TRC =====")
display(sinistre_trc.isnull().sum())
```

```python
missing_parc = parc_auto.isnull().mean() * 100
missing_auto = sinistre_auto.isnull().mean() * 100
missing_trc = sinistre_trc.isnull().mean() * 100

print("===== % VALEURS MANQUANTES PARC =====")
display(missing_parc.sort_values(ascending=False))

print("===== % VALEURS MANQUANTES AUTO =====")
display(missing_auto.sort_values(ascending=False))

print("===== % VALEURS MANQUANTES TRC =====")
display(missing_trc.sort_values(ascending=False))
```

```python
print("===== DOUBLONS =====")

print(f"Doublons parc auto : {parc_auto.duplicated().sum()}")

print(f"Doublons sinistre auto : {sinistre_auto.duplicated().sum()}")

print(f"Doublons sinistre TRC : {sinistre_trc.duplicated().sum()}")
```

```python
print("===== TYPES D'ENGINS =====")
display(parc_auto["type_engin"].value_counts())

print("===== ETAT GENERAL =====")
display(parc_auto["etat_general"].value_counts())

print("===== GRAVITE =====")
display(sinistre_auto["gravite"].value_counts())

print("===== TYPE TRAVAUX =====")
display(sinistre_trc["type_travaux"].value_counts())
```

## 5. Visualisations rapides

- histogrammes ;
- distributions ;
- fréquence des sinistres.

```python
# Histogramme 1
plt.figure(figsize=(10,5))

plt.hist(
    parc_auto["age_vehicule_ans"],
    bins=10
)

plt.title("Distribution de l'âge des véhicules")

plt.xlabel("Age véhicule")

plt.ylabel("Fréquence")

plt.show()
```

```python
# Histogramme 2
plt.figure(figsize=(10,5))

plt.hist(
    sinistre_auto["montant_indemnise_fcfa"],
    bins=20
)

plt.title("Distribution des montants indemnisés")

plt.xlabel("Montant indemnisé")

plt.ylabel("Fréquence")

plt.show()
```

```python
# Histogramme 3
plt.figure(figsize=(12,6))

sinistre_auto["type_accident"] \
    .value_counts() \
    .plot(kind='bar')

plt.title("Fréquence des types d'accidents")

plt.xlabel("Type accident")

plt.ylabel("Nombre")

plt.xticks(rotation=45)

plt.show()
```

```python
# Histogramme 4
plt.figure(figsize=(8,5))

sns.countplot(
    data=sinistre_auto,
    x="gravite"
)

plt.title("Répartition des niveaux de gravité")

plt.xticks(rotation=45)

plt.show()
```

```python
# Histogramme 5
cout_moyen = sinistre_auto.groupby("type_engin")[
    "montant_indemnise_fcfa"
].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))

cout_moyen.plot(kind='bar')

plt.title("Coût moyen par type d'engin")

plt.ylabel("Montant moyen")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    sinistre_auto.isnull(),
    cbar=False
)

plt.title("Heatmap des valeurs manquantes - Sinistre Auto")

plt.show()
```

```python
# Histogramme 6
plt.figure(figsize=(10,5))

sns.boxplot(
    x=sinistre_auto["montant_indemnise_fcfa"]
)

plt.title("Détection des valeurs aberrantes")

plt.show()
```

```python
sinistre_auto["date_accident"] = pd.to_datetime(
    sinistre_auto["date_accident"]
)

sinistre_auto["mois_accident"] = \
    sinistre_auto["date_accident"].dt.month

sinistres_par_mois = sinistre_auto.groupby(
    "mois_accident"
).size()

plt.figure(figsize=(10,5))

sinistres_par_mois.plot(marker='o')

plt.title("Evolution mensuelle des sinistres")

plt.xlabel("Mois")

plt.ylabel("Nombre de sinistres")

plt.show()
```

```python
# Matrice de correlation
variables_num = sinistre_auto.select_dtypes(
    include=np.number
)

plt.figure(figsize=(12,8))

sns.heatmap(
    variables_num.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Matrice de corrélation")

plt.show()
```
