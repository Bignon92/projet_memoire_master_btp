# Objectif

Nettoyer les données afin d’obtenir un dataset fiable.  

**Pertinence dans le mémoire**

Cette étape garantit :
- la qualité analytique ;
- la cohérence des résultats ;
- la fiabilité des modèles prédictifs.

```python
#Importation des libraires
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)

print("Librairies chargées")
```

```python
# Chargement des données
parc_auto = pd.read_csv(
    "../data/raw/parc_vehicules_porteo.csv"
)

sinistre_auto = pd.read_csv(
    "../data/raw/sinistres_auto_porteo_2022_2024.csv"
)

sinistre_trc = pd.read_csv(
    "../data/raw/sinistres_trc_porteo_2022_2024.csv"
)

print("Datasets chargés")
```

```python
# COPIES DE SÉCURITÉ
parc = parc_auto.copy()

auto = sinistre_auto.copy()

trc = sinistre_trc.copy()

print("Copies créées")
```

```python
# VÉRIFICATION DES DIMENSIONS
print(parc.shape)
print(auto.shape)
print(trc.shape)
```

## 1. Suppression des doublons

```python
parc = parc.drop_duplicates()

auto = auto.drop_duplicates()

trc = trc.drop_duplicates()

print("Doublons supprimés")
```

```python
# Vérification après suppression
print("Parc :", parc.duplicated().sum())

print("Auto :", auto.duplicated().sum())

print("TRC :", trc.duplicated().sum())
```

## 2. Gestion des valeurs manquantes

- moyenne ;
- médiane ;
- suppression ;
- valeurs par défaut.

```python
# STANDARDISATION DES NOMS DE COLONNES
parc.columns = parc.columns.str.lower().str.strip()

auto.columns = auto.columns.str.lower().str.strip()

trc.columns = trc.columns.str.lower().str.strip()

print("Colonnes standardisées")
```

```python
# AFFICHAGE DES COLONNES
print(parc.columns)

print(auto.columns)

print(trc.columns)
```

## 3. Correction des types

- dates ;
- numériques ;
- booléens.

```python
# CONVERSION DES DATES
auto["date_accident"] = pd.to_datetime(
    auto["date_accident"],
    errors="coerce"
)

trc["date_declaration"] = pd.to_datetime(
    trc["date_declaration"],
    errors="coerce"
)

print("Dates converties")
```

```python
# VÉRIFICATION TYPES
print(auto.dtypes)

print(trc.dtypes)
```

```python
# TRAITEMENT DES ESPACES
colonnes_objets_parc = parc.select_dtypes(
    include='object'
).columns

colonnes_objets_auto = auto.select_dtypes(
    include='object'
).columns

colonnes_objets_trc = trc.select_dtypes(
    include='object'
).columns

for col in colonnes_objets_parc:
    parc[col] = parc[col].str.strip()

for col in colonnes_objets_auto:
    auto[col] = auto[col].str.strip()

for col in colonnes_objets_trc:
    trc[col] = trc[col].str.strip()

print("Espaces supprimés")
```

## 4. Uniformisation des catégories

Exemple :
- Grave / grave / GRAVE
- Oui / OUI / yes

```python
#UNIFORMISATION MAJUSCULES/MINUSCULES
for col in colonnes_objets_parc:
    parc[col] = parc[col].str.upper()

for col in colonnes_objets_auto:
    auto[col] = auto[col].str.upper()

for col in colonnes_objets_trc:
    trc[col] = trc[col].str.upper()

print("Uniformisation effectuée")
```

```python
# Valeurs Manquantes
print(parc.isnull().sum())

print(auto.isnull().sum())

print(trc.isnull().sum())
```

## 5. Détection des anomalies

- montants négatifs ;
- âge incohérent ;
- délais impossibles.

```python
#IMPUTATION ÂGE VÉHICULE
median_age = parc["age_vehicule_ans"].median()

parc["age_vehicule_ans"] = parc[
    "age_vehicule_ans"
].fillna(median_age)

print("Valeurs manquantes âge traitées")
```

```python
# IMPUTATION KILOMÉTRAGE
median_km = parc[
    "km_parcourus_estimes"
].median()

parc["km_parcourus_estimes"] = parc[
    "km_parcourus_estimes"
].fillna(median_km)

print("Kilométrage imputé")
```

```python
# IMPUTATION ÉTAT GÉNÉRAL
mode_etat = parc["etat_general"].mode()[0]

parc["etat_general"] = parc[
    "etat_general"
].fillna(mode_etat)

print("Etat général imputé")
```

```python
# IMPUTATION MONTANTS
auto["montant_indemnise_fcfa"] = auto[
    "montant_indemnise_fcfa"
].fillna(0)

trc["montant_indemnise_fcfa"] = trc[
    "montant_indemnise_fcfa"
].fillna(0)

print("Montants imputés")
```

```python
# VÉRIFICATION DES VALEURS NÉGATIVES
print(
    auto[
        auto["montant_indemnise_fcfa"] < 0
    ].shape
)

print(
    trc[
        trc["montant_indemnise_fcfa"] < 0
    ].shape
)
```

```python
# CORRECTION VALEURS NÉGATIVES
auto.loc[
    auto["montant_indemnise_fcfa"] < 0,
    "montant_indemnise_fcfa"
] = 0

trc.loc[
    trc["montant_indemnise_fcfa"] < 0,
    "montant_indemnise_fcfa"
] = 0

print("Valeurs négatives corrigées")
```

```python
# SUPPRESSION DES ÂGES INCOHÉRENTS
parc = parc[
    (parc["age_vehicule_ans"] >= 0)
    &
    (parc["age_vehicule_ans"] <= 50)
]

print("Âges incohérents supprimés")
```

```python
# TRAITEMENT DES DÉLAIS
auto["delai_declaration_jours"] = auto[
    "delai_declaration_jours"
].fillna(0)

auto["delai_reglement_jours"] = auto[
    "delai_reglement_jours"
].fillna(0)

trc["delai_declaration_jours"] = trc[
    "delai_declaration_jours"
].fillna(0)

trc["delai_reglement_jours"] = trc[
    "delai_reglement_jours"
].fillna(0)

print("Délais nettoyés")
```

```python
# CORRECTION DES TYPES NUMÉRIQUES
colonnes_numeriques_auto = [

    "montant_declare_fcfa",
    "franchise_fcfa",
    "montant_indemnise_fcfa",
    "delai_declaration_jours",
    "delai_reglement_jours"

]

for col in colonnes_numeriques_auto:

    auto[col] = pd.to_numeric(
        auto[col],
        errors="coerce"
    )

print("Colonnes numériques converties")
```

```python
# VÉRIFICATION FINALE DES NULL
print(parc.isnull().sum())

print(auto.isnull().sum())

print(trc.isnull().sum())
```

```python
# CRÉATION COLONNE TYPE_SINISTRE
auto["type_sinistre"] = "AUTO"

trc["type_sinistre"] = "TRC"

print("Colonne type_sinistre créée")
```

```python
# UNIFORMISATION DES IMMATRICULATIONS
parc["immatriculation"] = parc[
    "immatriculation"
].str.replace(" ", "")

auto["immatriculation"] = auto[
    "immatriculation"
].str.replace(" ", "")

print("Immatriculations uniformisées")
```

```python
# CONTRÔLE DES CATÉGORIES
print(auto["gravite"].value_counts())

print(auto["responsabilite"].value_counts())

print(trc["nature_sinistre"].value_counts())
```

**EXPORT DES DONNÉES NETTOYÉES**

```python
parc.to_csv(
    "../data/interim/parc_auto_clean.csv",
    index=False
)

auto.to_csv(
    "../data/interim/sinistre_auto_clean.csv",
    index=False
)

trc.to_csv(
    "../data/interim/sinistre_trc_clean.csv",
    index=False
)

print("Datasets nettoyés exportés")
```

**RAPPORT FINAL DE NETTOYAGE**

```python
rapport_cleaning = pd.DataFrame({

    "Dataset": [
        "Parc Auto",
        "Sinistre Auto",
        "Sinistre TRC"
    ],

    "Nb_lignes": [
        parc.shape[0],
        auto.shape[0],
        trc.shape[0]
    ],

    "Nb_colonnes": [
        parc.shape[1],
        auto.shape[1],
        trc.shape[1]
    ]

})

rapport_cleaning.to_csv(

    "../reports/table/rapport_nettoyage.csv",
    index=False

)

print("Rapport nettoyage exporté")
```
